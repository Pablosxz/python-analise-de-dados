```python
import pandas as pd
import numpy as np
```

```python

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                   'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                       'B': ['B4', 'B5', 'B6', 'B7'],
                       'C': ['C4', 'C5', 'C6', 'C7'],
                       'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
   
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                     'C': ['C8', 'C9', 'C10', 'C11'],
                      'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
```

```python
dicionario = {'ID':[10,11,12,13,14,],'NOME':['Mario','Joana','Claudia','Lucas','Gabriel'],
              'CIDADE':['SÃO PAULO','RIO DE JANEIRO','SÃO PAULO','CAMPINAS','PORTO ALEGRE']}
dicionario2 = {'ID':[16,11,12,13,18,],'Experiência':['Junior','Senior','Pleno','Estagiário','Analista']}
```

```python
df4 = pd.DataFrame(data=dicionario)
df5 = pd.DataFrame(data=dicionario2)
```

```python
# Função Merge
```

```python
df4
```

```python
df5
```

```python
# Inner Merge
```

```python
merge_inner = pd.merge(df4,df5) # Junta apenas os que possuem o mesmo ID Horizontalmente
```

```python
merge_inner
```

```python
# Outer Merge
```

```python
merge_outer = pd.merge(df4,df5,'outer') # Junta todos os valores Horizontalmente (os que não possuem correspondentes em ambas as
# tabelas, são colocados como Nulo (NaN))
```

```python
merge_outer
```

```python
# Left Merge
```

```python
merge_left = pd.merge(df4,df5,how='left') # Junta os dados horizontalmente, mas utiliza apenas os IDs do DF da esquerda como principal
```

```python
merge_left
```

```python
# Right Merge
```

```python
merge_right = pd.merge(df4,df5,'right') # Junta os dados horizontalmente, mas utiliza apenas os IDs do DF da direita como principal
```

```python
merge_right
```

```python
# Função Join - É utilizado quando os Dataframes tem indices em comum
```

```python
join_esq = df4.join(df5,lsuffix=' X',rsuffix=' Y')
```

```python
join_esq
```
