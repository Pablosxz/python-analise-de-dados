```python
import pandas as pd
import numpy as np
```

```python
# Criação de um Dataframe baseado nos dados de uma pequena empresa, com as linhas contendo os meses que possuem 31 dias 
# e as colunas os dias do mês
```

```python
indices = ['Janeiro', 'Março', 'Maio', 'Julho', 'Agosto', 'Outubro','Dezembro']
```

```python
colunas = np.arange(1,32)
```

```python
colunas
```

```python
dados = np.random.randint(492,2050,size=(7,31)) # Gerando dados aleatórios para a suposta empres onde: 
# (valor mais baixo,valor mais alto,size=(linhas,colunas))
```

```python
dados
```

```python
df = pd.DataFrame(data=dados,index=indices,columns=colunas)
```

```python
df # Valor de venda de cada dia de cada mês
```

```python
df.loc['Janeiro',3] # Método localiza
```

```python
df[2] # Coluna 2
```

```python
df[[3,4,6]] # Lista de colunas
```

```python
# Quanto vendeu nos dias 23 de todos os meses?
```

```python
df[23]
```

```python
# Quanto vendeu nos dias 10, 20 e 30 de todos os meses?
```

```python
df[[10,20,30]]
```

```python
# Quanto vendeu no dia 15 de março?
```

```python
df[15]['Março'] # df [coluna][linha]

# Outra solução seria "df.loc['Março',15]"
```

```python
# Quanto vendeu nos meses de Janeiro, Março e Maio
```

```python
df[0:3] # df[inicial:final], com ":" trata apenas das linhas
```

```python
# Quanto vendeu nos meses de Julho, Agosto e Outubro nos dias 25 a 31?
```

```python
df.loc[['Julho','Agosto','Outubro'],np.arange(25,32)] # Localizar as linhas (meses) e colunas (dias) escolhidas.
```

```python
df.iloc[[3,4,5],np.arange(24,31)] # Outra Forma - Utilizando os indices para localizar 
```

```python
# Obs: "df.rename({atual:final},inplace=True)" serve para renomear alguma linha/coluna
```

```python
# Retornar um Dataframe apenas com os valores maiores que 1000
```

```python
df[df>1000]
```

```python
df[df>1000].fillna('-') # O método substitui os valors nulos (NaN) por "-"
```

```python
# Quais foram os valores maiores que 1000 dos dias 1 ao 5 de todos os meses?
```

```python
df[df>1000][:][np.arange(1,6)].fillna('-') # df[condição][linhaInicial:linhaFinal][colunas]
```

```python
# Em que meses os dias 10 e 15 foram maiores do que 1000?
```

```python
df[(df[10]>1000) & (df[22]>1500)] # Para retornar as linhas em que as condições forem atendidas: "df[(Condição1) & (Condição2)]"
```
