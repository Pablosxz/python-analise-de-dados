```python
# Criando uma coluna com o total de gols em cada partida e calculando a média de gols
```

```python
df['Total_de_gols'] = df['mandante_Placar'] + df['visitante_Placar']
```

```python
# Qual estado possui o maior numero de vitórias?
```

```python
df.head()
```

```python
# Criando a Coluna Estado_vencedor através de condições
```

```python
for index, row in df.iterrows(): # itera os valores de index (indice) e row (todos os valores de uma linha).
    if df.loc[index,'mandante_Placar'] > df.loc[index,'visitante_Placar']:
        df.loc[index,'Estado_vencedor'] = df.loc[index,'mandante_Estado']
    elif df.loc[index,'mandante_Placar'] < df.loc[index,'visitante_Placar']:
        df.loc[index,'Estado_vencedor'] = df.loc[index,'visitante_Estado']
    else:
        df.loc[index,'Estado_vencedor'] = '-'
```
